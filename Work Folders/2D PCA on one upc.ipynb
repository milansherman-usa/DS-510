```python
#create dataframe using the most common upc
df = df[df['upc'].isin([12000809960])]
df.shape
```

```python
df = df.set_index('7dmovement')
df.index.names = ['7dmovement']
df.head
```

```python
df= df.drop(['Unnamed'])
df.head(5)
```

```python
df= df.drop(['Unnamed'])
df.head(5)
```

```python
for col_name in df.columns:
    if(df[col_name].dtype == 'object'):
        df[col_name]= df[col_name].astype('category')
        df[col_name] = df[col_name].cat.codes
```

```python
df.dtypes
```

```python
import numpy as np
np.where(pd.isnull(df))
```

```python
from sklearn import preprocessing
data_scaled = pd.DataFrame(preprocessing.scale(df),columns = df.columns) 
```

```python
from sklearn.decomposition import PCA   
pca = PCA(n_components=2)
pca.fit_transform(data_scaled)
```

```python
print (pd.DataFrame(pca.components_,columns=data_scaled.columns,index = ['PC-1','PC-2']))
```

```python
existing_2d = pca.transform(df)
```

```python
df_2d = pd.DataFrame(existing_2d)
df_2d.index = df.index
df_2d.columns = ['PC1','PC2']
df_2d.head()
```

```python
%matplotlib inline
    
ax = df_2d.plot(kind='scatter', x='PC2', y='PC1', figsize=(16,8))
    
for i, upc in enumerate(df.index):
    ax.annotate(upc, (df_2d.iloc[i].PC2, df_2d.iloc[i].PC1))
```

```python
from sklearn.preprocessing import normalize
    
df_2d['demand_mean'] = pd.Series(df.mean(axis=1), index=df_2d.index)
demand_mean_max = df_2d['demand_mean'].max()
demand_mean_min = df_2d['demand_mean'].min()
demand_mean_scaled = (df_2d.demand_mean-demand_mean_min) / demand_mean_max
df_2d['demand_mean_scaled'] = pd.Series(demand_mean_scaled, index=df_2d.index) 
df_2d.head()
```

```python
df_2d.plot(
    kind='scatter', 
    x='PC2', 
    y='PC1', 
    s=df_2d['demand_mean_scaled']*100, figsize=(16,8))
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters = 10)
clusters = kmeans.fit(df)
```

```python
df_2d['cluster'] = pd.Series(clusters.labels_, index=df_2d.index)
```

```python
import numpy as np
    
df_2d.plot(
        kind='scatter',
        x='PC2',y='PC1',
        c=df_2d.cluster.astype(np.float), 
        figsize=(16,8),
        colormap = 'Set1')
```
